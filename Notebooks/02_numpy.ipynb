Générer un array de nombres aléatoires puis remplacer les valeurs plus petites que 0 par 0. Trier l’array obtenu


```python
# il s'agit de nombres issus d'une distribution normale centrée réduite
array6=np.random.randn(1000000)
```

```python
array6.mean()
```

```python
array6.var()
```

```python
array7=np.sort(np.where(array6<0,0,array6))
```

```python
array7.mean()
```

```python
# si on le fait en Python, on aura :
%%timeit
array8=np.empty(array6.shape[0])
for i in range(array6.shape[0]):
    if array6[i]<0:
        array8[i]=0
    else:
        array8[i]=array6[i]
```
